```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
```

```python
def datosAlmacenados(N):
    #N = el numero de datos que se van a tomar tipo int
    theta = ([i*np.pi for i in np.linspace(0,0.5,N)])#crea una lista convalores de 0 hasta pi/2 de menor a mayor
    condicionesIniciales = {'v0':2.34,'theta0':theta}#el diccionario
    datosR = [list(alturaMaxima(condicionesIniciales)),theta]#evalua la altura maxima para cada theta
    #print(datosR)
    return(datosR)
   
 
  
```

```python
def alturaMaxima(cI):
    #recibe un diccionario con una velocidad fija y un array de thetas
    g2 = 9.8 * 2 #la constante gravitacional multiplicada por 2
    v0 = cI['v0'] # la velocidad inicial
    thetas = list(cI['theta0'])#la lista de angulos 
    datos = list(map(lambda i: round((pow(v0,2)*pow(np.sin(i),2))/g2,3), thetas))#crea una lista con los valores de la altura maxima para cada theta
    return(datos)

    
```

```python
alturas,angulos = datosAlmacenados(30)
angulos = [ i/np.pi for i in angulos]
```

```python
max = 0.5
fig, ax = plt.subplots()
plt.scatter(angulos,alturas,s=20)
plt.title('Ymax en función del angulo')
plt.xlabel('angles(radians)',fontweight='bold', color = 'blue', fontsize='10', horizontalalignment='center')
plt.ylabel('max height m',fontweight='bold', color = 'blue', fontsize='10', horizontalalignment='center')
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=0.1))
plt.xlim(right=max+0.01)
plt.ylim(0,0.30)
plt.grid(True)
plt.savefig('figura')
```
