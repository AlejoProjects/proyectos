```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
datosnp = np.genfromtxt('datos-carro-1d.csv',skip_header=5,delimiter=',',missing_values=np.nan)
datos = pd.DataFrame(datosnp)
datos.columns = ['t','x','y','vx','vy','ax','ay']
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(figsize=(7,5))

axes.plot(datos['t'],datos['x'],marker = 'o',markerfacecolor = 'red',markersize = 4,label='x(t)')
axes.set_title('Posición en función del tiempo ')
axes.set_xlabel('tiempo')
axes.legend()
axes.set_ylabel('pos x')
plt.savefig('posicionVsTiempo')
```
